```python
class Human:
    #コンストラクタを定義する
    def __init__(self,name,age):
        self.name = name
        self.age= age

    #nameとageを標準出力するメソッドを追加
    def printinfo(self):
        print(f"Name:{self.name},Age:{self.age}")


#インスタンス化する
human = Human("侍太郎",36)

#printinfoメソッドを呼び出して属性を出力する
human.printinfo()
```
