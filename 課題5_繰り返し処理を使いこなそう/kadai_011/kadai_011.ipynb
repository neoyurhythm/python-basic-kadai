```python
planets =["水","金","地","火","木","土","天","海","冥"]

#配列planet_namesの要素を１つずつ順番に出力する
for planet in planets:
    print(planet)
```

```python
# リストの定義
planets = ["水", "金", "地", "火", "木", "土", "天", "海","冥"]
# インデックスの初期化
index = 0

# while文を使ってリストの要素を順に表示
while index < len(planets):
    print(planets[index])
    index = index + 1
```
