```python

def calculate_total_price(price,tax_rate = 0.10):
    #消費税額を計算する
    tax_amount = price * tax_rate

    #合計金額を計算する
    total_price = price + tax_amount

    #合計金額の返却
    return total_price

#関数を呼び出し、商品の金額と消費税率を渡す
price = 500
tax_rate = 0.10 #消費税10％
total = calculate_total_price(price,tax_rate)
print(f"{total}円")




```
